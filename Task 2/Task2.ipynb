```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
data = pd.read_csv('data.csv')
data.head(10)
```

```python
#заполнение NaN нулями
data = data.fillna(0)
```

```python
X = data.drop(columns = ['Итоговая оценка за экзамен по ТОЭ'])  
Y = data['Итоговая оценка за экзамен по ТОЭ']
```

```python
#Преобразование в числовой формат
X = pd.get_dummies(X, drop_first = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Модель регрессии
model = LinearRegression()
```

```python
#Обучение
model.fit(X_train, y_train)
```

```python
Y_th = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, Y_th)
r2 = r2_score(y_test, Y_th)
```

```python
print(f"Средняя абсолютная ошибка (MAE): {mse}")
print(f"Коэффициент детерминации (R^2): {r2}")
```
